```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from gensim import corpora, models 
import gensim
import os
from os import path
from time import sleep
```

```python
tokenizer = RegexpTokenizer(r'\w+')
en_stop = set(get_stop_words('en'))
import io
with io.open('result.txt', 'r', encoding='utf8', errors='replace') as f:
    texts = f.read()
    f.close
    
```

```python
tdm = []
en_stop_theo = ["thou", "us", "must", "since", "let", "anyone", "whole", "unless", "whole", "bring", "keep", "one", "things", "also", "says", "will", "hast", "every" "let", "may", "say", "said", "ye", "thee", "great", "even", "shall", "much", "now", "thy", "though", "nothing", "yet", "hath", "now", "upon", "saith", "many", "without", "see", "many", "unto", "another", "thing", "others", "s", "therefore", "thus", "indeed", "might", "rather", "know", "account", "tell", "come", "give", "can", "means", "saying", "words", "neither", "made"]

#Converted all the text to lowercase for uniform analysis
raw = texts.lower()
#Tokenized the text to individual terms and created the stop list
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in en_stop]
stopped_tokens_final = [j for j in stopped_tokens if not j in en_stop_theo]
tdm.append(stopped_tokens_final)

```

```python
dictionary = corpora.Dictionary(tdm)
corpus = [dictionary.doc2bow(i) for i in tdm]
sleep(3)
num_topics = 8
#Implemented the LdaModel
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=8, id2word = dictionary,  passes=4, alpha=[0.01]*num_topics, eta=[0.01]*len(dictionary.keys()))
print(ldamodel.print_topics(num_topics=5, num_words=7))
```
