```python
import re 
file = open("tokenize.txt", "r")
text = file.read()
message = re.split("\n\n",text)
y = range(len(message))
newm = []
for j in y:
  temp = message[j].replace("\n",' ')
  newm.append(temp)
    
k = range(len(newm))
new1 =[]
for u in k:
    temp1 = newm[u].replace("\t\t\t\t",' ')
    new1.append(temp1)
new1

o = range(len(new1))
new2 =[]
for y in o:
    temp2 = new1[y].replace("\t\t\t",' ')
    new2.append(temp2)
new2

o = range(len(new2))
new3 =[]
for u in o:
    temp3 = new2[u].replace("\t\t",' ')
    new3.append(temp3)
new3

o = range(len(new3))
new0 =[]
for u in o:
    temp4 = new3[u].replace("\t",' ')
    new0.append(temp4)
new0


o = range(len(new0))
new00 =[]
for u in o:
    temp5 = new0[u].replace("(",'')
    new00.append(temp5)
new00

o = range(len(new00))
new000 =[]
for u in o:
    temp6 = new00[u].replace(")",'')
    new000.append(temp6)
new000


o = range(len(new000))
new0001 =[]
for u in o:
    temp8 = new000[u].replace("[.]",'')
    new0001.append(temp8)
new0001
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from gensim import corpora, models 
import gensim
import os
from os import path
from time import sleep
```

```python
tokenizer = RegexpTokenizer(r'\w+')
en_stop = set(get_stop_words('en'))
import io
with io.open('result.txt', 'r', encoding='utf8', errors='replace') as f:
    texts = f.read()
    f.close
    
```

```python
tdm = []

#Converted all the text to lowercase for uniform analysis
raw = texts.lower()
#Tokenized the text to individual terms and created the stop list
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in en_stop]
tdm.append(stopped_tokens)

```

```python
dictionary = corpora.Dictonary(tdm)
corpus = [dictionary.doc2bow(i) for i in tdm]
sleep(3)
#Implemented the LdaModel
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=5, id2word = dictionary)
print(ldamodel.print_topics(num_topics=1, num_words=100))
```
