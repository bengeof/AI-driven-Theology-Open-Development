```python
import nltk.data
import codecs
import os

doc = codecs.open('result.txt', 'r')
content = doc.read()

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

f = open("tokenize.txt", "a")   
f.write ('\n\n'.join(tokenizer.tokenize(content)))
f.close()

```

```python
import re 
file = open("tokenize.txt", "r")
text = file.read()
message = re.split("\n\n",text)
y = range(len(message))
newm = []
for j in y:
  temp = message[j].replace("\n",' ')
  newm.append(temp)
    
k = range(len(newm))
new1 =[]
for u in k:
    temp1 = newm[u].replace("\t\t\t\t",' ')
    new1.append(temp1)
new1

o = range(len(new1))
new2 =[]
for y in o:
    temp2 = new1[y].replace("\t\t\t",' ')
    new2.append(temp2)
new2

o = range(len(new2))
new3 =[]
for u in o:
    temp3 = new2[u].replace("\t\t",' ')
    new3.append(temp3)
new3

o = range(len(new3))
new0 =[]
for u in o:
    temp4 = new3[u].replace("\t",' ')
    new0.append(temp4)
new0
```
