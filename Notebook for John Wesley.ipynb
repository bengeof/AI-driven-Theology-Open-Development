```python
from bs4 import BeautifulSoup

with open('sermonwesley2.xml', 'r') as f:

    contents = f.read()

    soup = BeautifulSoup(contents, 'lxml')

    tags = soup.find_all("p")
    
print(len(tags))
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

def a(text, brackets="()[]"):
    count = [0] * (len(brackets) // 2) # count open/close brackets
    saved_chars = []
    for character in text:
        for i, b in enumerate(brackets):
            if character == b: # found bracket
                kind, is_close = divmod(i, 2)
                count[kind] += (-1)**is_close # `+1`: open, `-1`: close
                if count[kind] < 0: # unbalanced bracket
                    count[kind] = 0  # keep it
                else:  # found bracket to remove
                    break
        else: # character is not a [balanced] bracket
            if not any(count): # outside brackets
                saved_chars.append(character)
    return ''.join(saved_chars)
```

```python
x = range(len(tags))
f = open("result.txt", "a")   
for i in x: 
    if hasNumbers(tags[i].text.strip()): 
        string = tags[i].text.strip()
        temp11 = string.replace("(",'')
        temp12 = temp11.replace(")",'')
        temp13 = a(temp12)
        temp14 = "".join(filter(lambda x: not x.isdigit(), temp13))
        f.write(temp14)
f.close()
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from gensim import corpora, models 
import gensim
import os
from os import path
from time import sleep
```

```python
tokenizer = RegexpTokenizer(r'\w+')
en_stop = set(get_stop_words('en'))
import io
with io.open('result1.txt', 'r', encoding='utf8', errors='replace') as f:
    texts = f.read()
    f.close
```

```python
tdm = []
en_stop_theo = ["thou", "answer", "end", "far", "ways", "whether", "works", "certain", "secondly", "hence", "something", "part", "consequently", "above", "abovei", "act", "according", "q", "hence", "wherefore", "end", "wherefore", "two", "stated", "first", "us", "must", "since", "let", "anyone", "whole", "unless", "whole", "bring", "keep", "one", "things", "also", "says", "will", "hast", "every" "let", "may", "say", "said", "ye", "thee", "great", "even", "shall", "much", "now", "thy", "though", "nothing", "yet", "hath", "now", "upon", "saith", "many", "without", "see", "many", "unto", "another", "thing", "others", "s", "therefore", "thus", "indeed", "every", "take", "yea", "day", "either", "well", "till", "never", "still", "ever", "might", "rather", "know", "account", "tell", "come", "give", "can", "means", "saying", "words", "neither", "made"]

#Converted all the text to lowercase for uniform analysis
raw = texts.lower()
#Tokenized the text to individual terms and created the stop list
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in en_stop]
stopped_tokens_final = [j for j in stopped_tokens if not j in en_stop_theo]
tdm.append(stopped_tokens_final)
```

```python
dictionary = corpora.Dictionary(tdm)
corpus = [dictionary.doc2bow(i) for i in tdm]
sleep(3)
num_topics = 10
#Implemented the LdaModel
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=num_topics, id2word = dictionary)
print(ldamodel.print_topics(num_topics=5, num_words=8))
```
